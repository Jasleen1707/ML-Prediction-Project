## WE CAN PREDICT HOW MANY MEDALS A COUNTRY WILL WIN IN THE OLYMPICS

```python
import pandas as pd
teams=pd.read_csv("teams.csv")
teams
```

```python
teams=teams[["team","country","year","athletes","age","prev_medals","medals"]]
teams
```

```python
teams.corr(numeric_only=True)["medals"]     #THE CORELATION BETWEEN THE POINTS
```

```python
#pip install seaborn
```

```python
import seaborn as sns
sns.lmplot(x="athletes",y="medals",data=teams,fit_reg=True, ci=None)
```

```python
sns.lmplot(x='age', y='medals' , data=teams, fit_reg=True,ci=None)  #NO REGRESSION
```

```python
teams[teams.isnull().any(axis=1)] #SHOWS THE NUMBER OF MISSING VALUE ROWS
```

```python
teams=teams.dropna()   #REMOVING  DATA 
teams
```

```python
train=teams[teams["year"] < 2012].copy()   #SPLIT THE DATA IN TRAINING AND TESTING SETS
test=teams[teams["year"] >= 2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
#pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
predictors = ["athletes", "prev_medals"]
reg.fit(train[predictors], train["medals"])
predictions = reg.predict(test[predictors])
predictions
```

```python
test["predictions"] = predictions  #ADDING PREDICTIONS COLUMN IN THE TABLE
test
```

```python
test.loc[test["predictions"] < 0,"predictions"] = 0
```

```python
test["predictions"]=test["predictions"].round()
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(test["medals"] ,test["predictions"])
error
```

```python
teams.describe()["medals"]  #to check error is below std
```

```python
test[test["team"]=="USA"]   #SAMPLE TESTING
```

```python
test[test["team"]=="IND"]
```

```python
errors=(test["medals"]-test["predictions"]).abs()
```

```python
errors
```

```python
error_by_team=errors.groupby(test["team"]).mean()
```

```python
error_by_team
```

```python
medals_by_team=test["medals"].groupby(test["team"]).mean()
```

```python
error_ratio=error_by_team/medals_by_team
```

```python
error_ratio
```

```python
error_ratio[~pd.isnull(error_ratio)] #DATA CLEANING
```
